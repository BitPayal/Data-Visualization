# Data Visualization #

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import titanic data from seaborn library
df = sns.load_dataset('titanic')
```

```python
df.plot(kind='scatter', x='age', y='fare')
```

```python
df.plot(kind='scatter', x='age', y='fare', color = 'red')
```

```python
df.head()
```

```python
df.groupby('sex')['fare'].mean().plot(kind='bar')
plt.ylabel('Average Fare')
plt.xlabel('Gender')
```

```python
# seaborn plots

# bar plot using seaborn on df
sns.barplot(data = df, x= 'survived', y = 'fare')
```

```python
sns.scatterplot(data = df, y= 'fare', x = 'age', hue='class')
```

```python
# make scatter plot via plotly
import plotly.express as px

fig = px.scatter(df, x='age', y='fare', color='class')
fig.show()
```

```python
import plotly.graph_objects as go

import pandas as pd

# Read data from a csv
z_data = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/api_docs/mt_bruno_elevation.csv')

fig = go.Figure(data=[go.Surface(z=z_data.values)])

fig.update_layout(title='Mt Bruno Elevation', autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))

fig.show()
```
